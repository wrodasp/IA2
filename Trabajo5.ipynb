## A) Práctica 1

### 1. Solución manual

$f(x) = \frac{\cos(x)}{x}$

#### Iteración 1: 

##### - Etapa 0:

${x}^{(0)}=2,\:k=0$

##### - Etapa 1:

$\frac{\delta f(x)}{\delta x} = \frac{(x \cdot sin(x)+cos(x)}{x^{2}}$

$\left \| \nabla f(x^{(0)}) \right \| = -\frac{(2) \cdot sin(2)+cos(2)}{{2}^{2}}$

$\left \| \nabla f(x^{(0)}) \right \| = -0.35$

##### - Etapa 2:

$\delta^{(0)} = -\nabla f(x^{(0)})$

$\delta^{(0)} = -(-0.35)$

$\delta^{(0)} = 0.35$


$\min g_{x^{(0)}, \delta^{(0)}} = f(x^{(0)} + \alpha \cdot \delta^{(0)}) - f(x^{(0)})$

$\min g_{x^{(0)}, \delta^{(0)}} = f(2 + \alpha \cdot 0.35) - f(2)$

$\min g_{x^{(0)}, \delta^{(0)}} = \frac{cos(2 + 0.35 \alpha)}{2 + 0.35 \alpha} - \frac{cos(2)}{2}$

$\alpha = -2.35$


$x^{(1)} = x^{(0)} + \alpha \cdot \delta^{(0)}$

$x^{(1)} = 2 + (-2.35) \cdot (0.35)$

$x^{(1)} = 1.17$

#### Iteración 2:

##### - Etapa 0:

${x}^{(1)}=1.17,\:k=1$

##### - Etapa 1:

$\frac{\delta f(x)}{\delta x} = \frac{(x \cdot sin(x)+cos(x)}{x^{2}}$

$\left \| \nabla f(x^{(1)}) \right \| = -\frac{(1.17) \cdot sin(1.17)+cos(1.17)}{{1.17}^{2}}$

$\left \| \nabla f(x^{(1)}) \right \| = -1.06$

##### - Etapa 2:

$\delta^{(1)} = -\nabla f(x^{(1)})$

$\delta^{(1)} = -(-1.06)$

$\delta^{(1)} = 1.06$


$\min g_{x^{(1)}, \delta^{(1)}} = f(x^{(1)} + \alpha \cdot \delta^{(1)}) - f(x^{(1)})$

$\min g_{x^{(1)}, \delta^{(1)}} = f(1.17 + \alpha \cdot 1.06) - f(1.17)$

$\min g_{x^{(1)}, \delta^{(1)}} = \frac{cos(1.17 + 1.06 \alpha)}{1.17 + 1.06 \alpha} - \frac{cos(1.17)}{1.17}$

$\alpha = -2.23$


$x^{(2)} = x^{(1)} + \alpha \cdot \delta^{(1)}$

$x^{(2)} = 1.17 + (-2.23) \cdot (1.06)$

$x^{(2)} = -1.20$

#### Iteración 3:

##### - Etapa 0:

${x}^{(2)}=-1.20,\:k=2$

##### - Etapa 1:

$\frac{\delta f(x)}{\delta x} = \frac{(x \cdot sin(x)+cos(x)}{x^{2}}$

$\left \| \nabla f(x^{(2)}) \right \| = -\frac{(-1.2) \cdot sin(-1.2)+cos(-1.2)}{{-1.2}^{2}}$

$\left \| \nabla f(x^{(2)}) \right \| = -1.02$

##### - Etapa 2:

$\delta^{(2)} = -\nabla f(x^{(2)})$

$\delta^{(2)} = -(-1.02)$

$\delta^{(2)} = 1.02$


$\min g_{x^{(2)}, \delta^{(2)}} = f(x^{(2)} + \alpha \cdot \delta^{(2)}) - f(x^{(2)})$

$\min g_{x^{(2)}, \delta^{(2)}} = f(2 + \alpha \cdot 1.02) - f(2)$

$\min g_{x^{(2)}, \delta^{(2)}} = \frac{cos(-1.2 + 1.02 \alpha)}{-1.2 + 1.02 \alpha} - \frac{cos(-1.2)}{2}$

$\alpha = 0.18$


$x^{(3)} = x^{(0)} + \alpha \cdot \delta^{(2)}$

$x^{(3)} = -1.2 + (0.18) \cdot (1.02)$

$x^{(3)} = -1.01$

### 2) Implementación con Sympy

```python
#Importación de las librerias necesarias
import matplotlib.pyplot as pp
import sympy as sp
import numpy as np
%matplotlib inline

#Definición de la ecuación 
variable, alfa = sp.symbols('x, alfa')
ecuacion = sp.cos(variable) / variable

#Definición de funciones python para facilitar la solución
def resolver_ecuacion_derivada(x):
    derivada = sp.diff(ecuacion, variable)
    return derivada.subs(variable, x).evalf()

def graficar_progreso(x = None):
    rango = np.arange(-10, 10, 0.9)
    ecuacion = lambda x: np.cos(x) / x
    if x != None:
        pp.scatter(x, ecuacion(x), color = 'red')
    pp.plot(rango, ecuacion(rango), color = 'green')
    pp.show()

#Inicio del algoritmo de decenso por gradiente
#Etapa 0: Definición de (x) inicial y (k) = 0
x = 2

print('- Grafica de x =', 2, 'en k =', 0)
graficar_progreso(x)

for k in range(1, 4):
    #Etapa 1: Cálculo de la magnitud
    magnitud = resolver_ecuacion_derivada(x)

    if magnitud == 0:
        break
    else:
        #Etapa 2: Cálculo de la dirección y minimizamos la funcion para encontrar el nuevo (x)
        delta = -magnitud
        
        minima = ecuacion.subs(variable, x + alfa + delta) - ecuacion.subs(variable, x)
        minima = (sp.solveset(minima, alfa))
        
        a = min(minima.args[1])
        
        x = x + (float(a) * delta)
        
        print('- Grafica de x =', x, 'en k =', k)
        graficar_progreso(float(x))
```

## B) Práctica 2

### 1. Solución manual

$f(x,y) = x \cdot y^{2} - x^{2} \cdot y$



#### Iteración 1: 

##### - Etapa 0:

${(x,y)}^{(0)}=(1,2),\:k=0$

##### - Etapa 1:

$\frac{\delta f(x,y)}{\delta x} = y \cdot (y-2x)$

$\left \| \nabla f(x^{(0)}) \right \| = 2 \cdot (2-2(1))$

$\left \| \nabla f(x^{(0)}) \right \| = 0$


$\frac{\delta f(x,y)}{\delta y} = -x \cdot (x - 2y)$

$\left \| \nabla f(y^{(0)}) \right \| = -1 \cdot (1 - 2(2))$

$\left \| \nabla f(y^{(0)}) \right \| = 3$

##### - Etapa 2:

$\min g_{(x,y)^{(0)}, \delta^{(0)}} = -\nabla f((x,y)^{(0)} - \alpha \cdot \nabla f(x,y)^{(0)}) * f((x,y)^{(0)})^{T}$

$\min g_{(x,y)^{(0)}, \delta^{(0)}} = -\nabla f((1,2) - \alpha \cdot (0,3)) * (0,3)^{T}$

$\min g_{(x,y)^{(0)}, \delta^{(0)}} = -\nabla f(1, 2-3\alpha) * (0,3)^{T}$

$\min g_{(x,y)^{(0)}, \delta^{(0)}} = -((2-3\alpha) \cdot ((2-3\alpha)-2), -1 \cdot (-1 - 2(2-3\alpha))) * (0,3)^{T}$

$\alpha = 0.28$


$x^{(1)} = x^{(0)} - \alpha \cdot \delta f(x)^{(0)}$

$x^{(1)} = 1 - (0.28) \cdot (0)$

$x^{(1)} = 1$

$y^{(1)} = y^{(0)} - \alpha \cdot \delta f(y)^{(0)}$

$y^{(1)} = 2 - (0.28) \cdot (3)$

$y^{(1)} = 1.16$

#### Iteración 2: 

##### - Etapa 0:

${(x,y)}^{(1)}=(1,1.16),\:k=1$

##### - Etapa 1:

$\frac{\delta f(x,y)}{\delta x} = y \cdot (y-2x)$

$\left \| \nabla f(x^{(1)}) \right \| = 1.16 \cdot (1.16-2(1))$

$\left \| \nabla f(x^{(1)}) \right \| = -0.98$


$\frac{\delta f(x,y)}{\delta y} = -x \cdot (x - 2y)$

$\left \| \nabla f(y^{(1)}) \right \| = -1 \cdot (1 - 2(1.16))$

$\left \| \nabla f(y^{(1)}) \right \| = 1.26$

##### - Etapa 2:

$\min g_{(x,y)^{(1)}, \delta^{(1)}} = -\nabla f((x,y)^{(1)} - \alpha \cdot \nabla f(x,y)^{(1)}) * f((x,y)^{(0)})^{T}$

$\min g_{(x,y)^{(1)}, \delta^{(1)}} = -\nabla f((1,1.16) - \alpha \cdot (-0.98,1.26)) * (-0.98,1.26)^{T}$

$\min g_{(x,y)^{(1)}, \delta^{(1)}} = -\nabla f(1 + 0.98\alpha, 1.16-1.26\alpha) * (-0.98,1.26)^{T}$

$\min g_{(x,y)^{(1)}, \delta^{(1)}} = -((1.16-1.26\alpha) \cdot ((1.16-1.26\alpha)-2(1 + 0.98\alpha)) , -(1 + 0.98\alpha) \cdot ((1 + 0.98\alpha) - 2(1.16-1.26\alpha))) * (-0.98,1.26)^{T}$

$\alpha = 1.45$


$x^{(2)} = x^{(1)} - \alpha \cdot \delta f(x)^{(1)}$

$x^{(2)} = 1 - (1.45x10^{-16}) \cdot (-0.98)$

$x^{(2)} = 1$

$y^{(2)} = y^{(1)} - \alpha \cdot \delta f(y)^{(1)}$

$y^{(2)} = 1.16 - (1.45x10^{-16}) \cdot (1.26)$

$y^{(2)} = 1.16$

Como se puede observar, para k=2 los valores obtenidos para (x,y) son iguales a los valores en k=1, por lo tanto, se determina que la solución fue encontrada en dos iteraciones.

```python
#Definición de la ecuación 
var_x, var_y, alfa = sp.symbols('x, y, alfa')
ecuacion = (var_x * var_y**2) - (var_x**2 * var_y)

#Definición de funciones python para facilitar la solución
def resolver_ecuacion_base(x,y):
    return ecuacion.subs({var_x: x, var_y: y}).evalf()

def graficar_progreso(x, y):
    rango_x = np.arange(-10, 10, 1)
    rango_y = np.arange(-10, 10, 1)
    
    X, Y = np.meshgrid(rango_x, rango_y) 
    
    ecuacion = lambda x, y: np.dot(x, y**2) - np.dot(x**2, y)
    
    grafica = pp.figure()
    grafica.set_size_inches(5,5)
    ejes = pp.axes(projection = '3d')
    ejes.scatter(x, y, ecuacion(x, y), color = 'red', linewidths=5)
    ejes.plot_wireframe(X, Y, ecuacion(X, Y), color = 'green')

#Inicio del algoritmo de decenso por gradiente
#Etapa 0: Definición de (x,y) iniciales y (k) = 0
x = 1
y = 2

derivada_x = sp.diff(ecuacion, var_x)
derivada_y = sp.diff(ecuacion, var_y)

print('1. Grafica de (x, y) = (1, 2) en k = 0')
graficar_progreso(x, y)

for k in range(1, 2):
    #Etapa 1: Cálculo de las magnitudes
    
    dx = derivada_x.subs({var_x: x, var_y: y}).evalf()
    dy = derivada_y.subs({var_x: x, var_y: y}).evalf()

    if dx == 0 and dy == 0:
        break
    else:
        #Etapa 3: Cálculo de los nuevos puntos
        termino_1 = derivada_x.subs({var_x: x - (alfa * dx), var_y: y - (alfa * dy)})
        termino_2 = derivada_y.subs({var_x: x - (alfa * dx), var_y: y - (alfa * dy)})

        matriz_1 = -sp.Matrix([[termino_1, termino_2]])
        matriz_2 = sp.Matrix([1.08, 0.84])

        minima = (matriz_1 * matriz_2)
        
        a = min(sp.solve(minima, alfa))
        
        x = x - (a[0] * dx)
        y = y - (a[0] * dy)
        
        print('2. Grafica de (x, y) = (', x, ',', y,') en k =', k)
        graficar_progreso(float(x), float(y))

#Nota: El gráfico se ve como un plano porque np.meshgrid() crea las matrices de forma que los puntos de (x) y (y)
#      se correspondan correctamente, es decir que x0 y y0 son iguales y por lo tanto debido a la ecuación
#      propuesta f(x, y) = x*(y^2) - (x^2)*y el resultado (z) sera igual a 0 para todos los puntos. La función
#      normalmente deberia verse como en el gráfico del cálculo manual.
```

## Teoria - Funciones cóncavas y puntos estacionarios

#### 1) Funciones cóncavas

Una función es cóncava cuando dados dos puntos cualesquiera en el dominio de la función, el segmento que los une queda por debajo de la curva. Presenta su concavidad hacia abajo. Una función cóncava es lo opuesto de una función convexa.

#### 2) Puntos estacionarios

Un punto estacionario de una función de una variable real: es un número donde la derivada de es cero. Si la función es derivable y tiene un extremo local en un punto, ese punto estará entre sus puntos estacionarios.
