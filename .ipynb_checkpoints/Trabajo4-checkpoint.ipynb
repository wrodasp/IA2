##  A) Práctica ANN-0

```python
#Importación de las librerias
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as pp
import numpy as np
import random
import copy

%matplotlib inline

#Entradas y salidas (Sin Hot-Encoding)
x = np.array([[0,0],[0,1],[1,0],[1,1]])
d = np.array([0,1,1,0])

#Gráfica de los datos
pp.scatter(x[:,0], x[:,1], color = ['blue' if i == 1 else 'red' for i in d])
pp.grid(True)
pp.show()
```

```python
# Modificar las salidas deseadas para representarlas con 2 valores binarios
# Valores: 01 = 0 y 10 = 1
d = np.array([[0,1],[1,0],[1,0],[0,1]])

# Modificar los parametros de la Red Neuronal. Sustituir None por el valor correspondiente
neuronas_capa_oculta = 4
neuronas_capa_salida = 1

mlp = MLPClassifier(
    solver = 'lbfgs', 
    activation ='logistic', 
    alpha = 1e-4, 
    tol = 1e-15, 
    max_iter = 10000, 
    hidden_layer_sizes=(neuronas_capa_oculta, neuronas_capa_salida)
)

# Realizamos el proceso de entrenamiento con la nueva representacion de la salida. 
mlp.fit(x,d)

# Mostramos los pesos entre la entrada y la capa oculta
print('Pesos W^(0): \n:', mlp.coefs_[0])

# Mostramos los pesos entre la capa oculta y la capa de salida
print('\nPesos W^(1): \n:', mlp.coefs_[1])

# Probamos si la red devuelve valores apropiados de acuerdo a las entradas (test):
for entrada in x:
    print('\nPrueba con {',
          '|'.join([str(i) for i in entrada]),'} => ',
          mlp.predict(entrada.reshape(1,-1))
    )
```

## B) Práctica ANN-1

```python
#Definción de la ecuación a tratar
ecuacion = lambda x: np.cos(x) + x

#Creación de los puntos de la ecuación
x = np.arange(-5, 5, 0.01)
y = ecuacion(x)

#Generación de 1000 puntos aleatorios
a = copy.copy(x)
random.shuffle(a)

b = [aux + random.uniform(-1, 1) for aux in y]

#Clasificación de los puntos por clases
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(len(a)):
    if b[i] >= ecuacion(a[i]):
        x1.append([a[i], b[i]])
        y1.append(0)
    else:
        x2.append([a[i], b[i]])
        y2.append(1)

x1 = np.array(x1)
x2 = np.array(x2)
y1 = np.array(y1)
y2 = np.array(y2)

#Grafica de la ecuación y los 1000 puntos aleatorios
pp.plot(x, y, color = 'b', lw = 5.0)
pp.scatter(x1[:,0], x1[:,1], color = 'r')
pp.scatter(x2[:,0], x2[:,1], color = 'g')
pp.grid(True)

print('Cantidad de puntos: ', len(x))
pp.show()
```

```python
#División de los datos en entrenamiento y prueba (80%, 20%)
datos = np.vstack((x1, x2))
y_deseadas = np.append(y1, y2)

x_train, x_test, y_train, y_test = train_test_split(
    datos,
    y_deseadas,
    test_size = 0.2,
    stratify = y_deseadas
)

#Creación y entrenamiento de la red neuronal
red = MLPClassifier(
    solver = 'lbfgs', 
    activation ='logistic', 
    alpha = 1e-4,
    tol = 1e-15,
    max_iter = 10000, 
    hidden_layer_sizes=(2, 1) #(N. capa oculta = 4 y N. capa salida = 1)
)

red.fit(x_train, y_train)
matriz_confusion = np.array([[0,0], [0,0]])

for i in range(0, len(x_test)):
    if y_test[i] == 0:
        if red.predict(x_test[i].reshape(1, -1)) == 0:
            matriz_confusion[0][0] += 1
        else:
            matriz_confusion[0][1] += 1
    else:
        if red.predict(x_test[i].reshape(1, -1)) == 1:
            matriz_confusion[1][0] += 1
        else:
            matriz_confusion[1][1] += 1

precision = (matriz_confusion[0][0] + matriz_confusion[1][1]) / len(x_test)

print('1) Prueba de predicción de 10 muestras:\n')
for i in range(26, 36):
    y_calculada = red.predict(x_test[i].reshape(1, -1))
    print('Patron: ', x_test[i], ', "Y" esperada: ', y_test[i], ', "Y" calculada: ', y_calculada)
    
print('\n2)Matriz de confusión:\n')
print(matriz_confusion)

print('\nPrecisión de la red: ', precision)
```
